# Polynomial Regression

Polynomial regression is an extension of linear regression, where the relationship between the independent variable x and the dependent variable
y is modeled as an nth degree polynomial. This allows for a more flexible fit to data that might not be well-represented by a straight line.

## 1. Generate Synthetic Data

We'll create a dataset with a non-linear relationship between the independent variable x and the dependent variable y.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the random seed for reproducibility
np.random.seed(0)
```

```python
# Generate random data points
x = np.random.rand(100, 1) * 10  # Independent variable
y = 2 * x**2 + 3 * x + np.random.randn(100, 1) * 10  # Dependent variable
```

```python
# Scatter plot of the data
plt.scatter(x, y, color='blue')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of Synthetic Data")
plt.show()
```

## 2. Transform the Data to Include Polynomial Features

We'll use PolynomialFeatures from Scikit-learn to transform the input features into polynomial features.

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Transform the features to polynomial features of degree 2
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
```

```python
# Display the transformed features
print("Original feature shape:", x.shape)
print("Transformed feature shape:", x_poly.shape)
```

## 3. Fit a Polynomial Regression Model

We'll use the transformed polynomial features to fit a linear regression model.

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Fit the model to the polynomial features
model.fit(x_poly, y)
```

```python

# Predict using the polynomial model
y_pred = model.predict(x_poly)
```

## 4. Visualize the Results

We'll plot the original data points and the polynomial regression curve.

```python
# Sort the data points for plotting
sorted_idx = np.argsort(x[:, 0])
x_sorted = x[sorted_idx]
y_sorted = y_pred[sorted_idx]
```

```python
# Plot the results
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x_sorted, y_sorted, color='red', label='Polynomial regression curve')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```

## 5. Evaluate the Model

We'll evaluate the model's performance using Mean Squared Error (MSE) and R-squared (R^2).

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Calculate evaluation metrics
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
